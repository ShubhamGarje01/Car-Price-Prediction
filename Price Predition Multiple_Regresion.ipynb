## check using Model Validation Techniques

## 1.Colinearity Check
## 2.Residuals

```python
# 1.Colinearity Check
# formula : VIF = 1/(1-Rsquare)

rsq_age = smf.ols("Age ~ KM + HP + CC + Doors + Gears + Q_Tax + Weight", data = car2).fit().rsquared
vif_age = 1/(1 - rsq_age)

rsq_km = smf.ols("KM ~ Age + HP + CC + Doors + Gears + Q_Tax + Weight", data = car2).fit().rsquared
vif_age = 1/(1 - rsq_km)

rsq_hp = smf.ols("HP ~ Age + KM + Price + CC + Doors + Gears + Q_Tax + Weight", data = car2).fit().rsquared
vif_hp = 1/(1 - rsq_hp)

rsq_cc = smf.ols("CC ~ Age + KM + HP + Price + Doors + Gears + Q_Tax + Weight", data = car2).fit().rsquared
vif_cc = 1/(1 - rsq_cc)

rsq_doors = smf.ols("Doors ~ Age + KM + HP + CC + Price + Gears + Q_Tax + Weight", data = car2).fit().rsquared
vif_doors = 1/(1 - rsq_doors)

rsq_gears = smf.ols("Gears ~ Age + KM + HP + CC + Doors + Price + Q_Tax + Weight", data = car2).fit().rsquared
vif_gears = 1/(1 - rsq_gears)

rsq_qtax = smf.ols("Q_Tax ~ Age + KM + HP + CC + Doors + Gears + Price + Weight", data = car2).fit().rsquared
vif_qtax = 1/(1 - rsq_qtax)

rsq_weight = smf.ols("Weight ~ Age + KM + HP + CC + Doors + Gears + Q_Tax + Price", data = car2).fit().rsquared
vif_weight = 1/(1 - rsq_weight)
```
